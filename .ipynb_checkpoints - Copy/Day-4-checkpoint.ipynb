```python
# Function For Prime Numbers

def Is_Prime(n):
    cnt=0
    for i in range(1,n+1):
        if n%i==0:
            cnt+=1
    if cnt==2:
        return True
    else:
        return False
start=int(input("Enter start range: "))
end=int(input("Enter range: "))
prime_list=[i for i in range(start,end+1) if Is_Prime(i)]
print(len(prime_list))
print(prime_list,end=' ')
```
